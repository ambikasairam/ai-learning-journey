Learning Numpy

NumPy teaches you to think in arrays, not loops.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
a

#2D array
b = np.array([[1, 2, 3],
              [4, 5, 6]])
b

c = np.array([[1,2,3,4],[5,6,7,8]])

#AI models mostly operate on 2D+ arrays.

#array properties
a.shape      # dimensions
a.ndim       # number of dimensions
a.size       # total elements
a.dtype
print("shape is",b.shape)   # (2, 3)
print("dim is",b.ndim)
print("size",b.size)
print("type is ", b.dtype)

# Create a 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix) # Outputs the matrix

# Access an element
element = matrix[1, 2]
print(element) # Outputs: 6

#creating common arrays
np.zeros((3, 4))      # matrix of zeros
np.ones((2, 2))       # matrix of ones


#reshaping arrays - Create a 1D array
array = np.array([1, 2, 3, 4, 5, 6])
print("original array", array)

# Reshape the array to 2x3
reshaped_array = array.reshape(2, 3)
print("Reshaped array",reshaped_array) # Outputs a reshaped 2D array


```

```python
#slicing and dicing
a = np.array([1, 2, 3, 4])

#2D array
b = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a[0])
print(a[1:3])
print(b[0, 1]  )  # row 0, column 1
print(b[:, 1] )   # all rows, column 1
print(b[1, :] )   # row 1
```

```python
import numpy as np
# arange uses step size
arange_result = np.arange(0, 10, 2)
print("arange(0, 10, 2):", arange_result)
# Output: [0 2 4 6 8]

# linspace uses number of elements
linspace_result = np.linspace(0, 10, 2)
print("linspace(0, 10, 2):", linspace_result)
# Output: [0. 10.]

# More comparable example
arange_ex = np.arange(0, 10, 2.5)
print("\narange(0, 10, 2.5):", arange_ex)
# Output: [0. 2.5 5. 7.5]

linspace_ex = np.linspace(0, 10, 5)
print("linspace(0, 10, 5):", linspace_ex)
# Output: [0. 2.5 5. 7.5 10.]
```

```python
#Basic Copy
import numpy as np

arr = np.array([1, 2, 3])
copy_arr = np.copy(arr)
copy_arr[0] = 99

print(arr)      # Output: [1, 2, 3]
print(copy_arr) # Output: [99, 2, 3]
```

```python
#Basic View

import numpy as np

arr = np.array([1, 2, 3])
print(arr[0:3])
view_arr = arr.view()
view_arr[0] = 99

print(arr)      # Output: [99, 2, 3]
print(view_arr) # Output: [99, 2, 3]
```

```python
#slicing

import numpy as np

arr = np.array([[1, 2], [3, 4], [5, 6]])
slice_copy = np.copy(arr[0:2])
print("slice_copy", slice_copy)
slice_view = arr[0:2].view()
print("slice_view", slice_view)

slice_copy[0, 0] = 99
slice_view[0, 0] = 88

print("arr",arr)         # Output: [[ 1  2]
                   #          [88  4]
                   #          [ 5  6]]
print("slice_copy",slice_copy)  # Output: [[99  2]
                   #          [ 3  4]]
print("slice_view",slice_view)  # Output: [[88  2]
                   #          [ 3  4]]
```
